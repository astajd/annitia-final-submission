# 01 — Results review
Quick-look at the per-experiment summaries written by `python -m src.run_phase1`. All logic lives in `src/`.

```python
%load_ext autoreload
%autoreload 2
import sys, pathlib, json
ROOT = pathlib.Path.cwd().parent
if str(ROOT) not in sys.path: sys.path.insert(0, str(ROOT))

import pandas as pd
from src import config as cfg
out = sorted(cfg.EXPERIMENT_OUTPUTS.iterdir())
rows = []
for d in out:
    cv = d / 'cv_summary.json'
    cfgj = d / 'config.json'
    if cv.exists() and cfgj.exists():
        rows.append({**json.loads(cfgj.read_text()), **json.loads(cv.read_text()), 'dir': d.name})
df = pd.DataFrame(rows)
df[['name','feature_set','death_target_mode','cindex_hepatic_mean','cindex_death_mean','score_mean','score_std']].sort_values('score_mean', ascending=False)
```
